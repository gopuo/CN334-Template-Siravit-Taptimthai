  - ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô prompt ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö LLM ‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏£‡∏±‡∏ö 10 tuples (‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏≠‡∏á) ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á query ‡∏à‡∏£‡∏¥‡∏á‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡πÉ‡∏ä‡πâ‡∏ó‡∏î‡∏™‡∏≠‡∏ö ‡πÉ‡∏´‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á 3 queries ‡∏ï‡πà‡∏≠ 1 tuple. ‡∏£‡∏ß‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î 30 ‡∏≠‡∏±‡∏ô
  - ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏ô‡∏±‡πâ‡∏ô‡πÉ‡∏´‡πâ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ß‡πà‡∏≤ ‡∏ó‡∏µ‡πà‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏ô‡∏±‡πâ‡∏ô ‡∏≠‡∏±‡∏ô‡∏ô‡∏±‡πâ‡∏ô‡∏™‡∏°‡∏à‡∏£‡∏¥‡∏á‡πÅ‡∏•‡∏∞‡∏Ñ‡∏¥‡∏î‡∏ß‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏≥‡∏ñ‡∏≤‡∏°‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏õ‡πÑ‡∏î‡πâ‡∏ö‡πâ‡∏≤‡∏á ‡πÅ‡∏•‡∏∞‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏°‡∏≤‡πÉ‡∏™‡πà‡πÉ‡∏ô dataset ‡πÑ‡∏ß‡πâ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
  

**‡∏Ñ‡∏≥‡πÅ‡∏ô‡∏∞‡∏ô‡∏≥**: ‡∏î‡∏π‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏à‡∏≤‡∏Å slide ‡πÑ‡∏î‡πâ

**Answer**: ‡∏ï‡∏≠‡∏ö Key Dimensions ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢ ‡∏î‡πâ‡∏≤‡∏ô‡∏•‡πà‡∏≤‡∏á

1.cuisine_type ‡∏£‡∏∞‡∏ö‡∏∏‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏à‡∏≤‡∏Å‡∏õ‡∏£‡∏∞‡πÄ‡∏ó‡∏®‡∏´‡∏£‡∏∑‡∏≠‡∏™‡πÑ‡∏ï‡∏•‡πå‡πÄ‡∏â‡∏û‡∏≤‡∏∞

2.dietary_restriction ‡∏Ç‡πâ‡∏≠‡∏à‡∏≥‡∏Å‡∏±‡∏î‡∏î‡πâ‡∏≤‡∏ô‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç‡∏ï‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°‡πÅ‡∏•‡∏∞‡∏Ñ‡∏ß‡∏≤‡∏°‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢‡∏Ç‡∏≠‡∏á‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ ‡πÄ‡∏ä‡πà‡∏ô ‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏ö‡∏≤‡∏á‡∏Ñ‡∏ô‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏°‡∏±‡∏á‡∏™‡∏ß‡∏¥‡∏£‡∏±‡∏ï‡∏¥‡∏´‡∏£‡∏∑‡∏≠‡∏õ‡∏£‡∏≤‡∏®‡∏à‡∏≤‡∏Å‡∏Å‡∏•‡∏π‡πÄ‡∏ï‡∏ô

3.meal_type ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏°‡∏∑‡πâ‡∏≠‡∏≠‡∏≤‡∏´‡∏≤‡∏£ (‡πÄ‡∏ä‡πà‡∏ô breakfast, lunch, dinner)





````python
tuple_gen_prompt = """
You are generating test data for a Recipe Recommendation Bot.

Dimensions:
1. cuisine_type: [Thai, Italian, Japanese]
2. dietary_restriction: [vegetarian, gluten-free, halal]
3. meal_type: [breakfast, lunch, dinner]

Task:
- Generate **15 unique tuples**
- Each tuple must contain exactly (cuisine_type, dietary_restriction, meal_type)
- Ensure no duplicates and cover variation across all dimensions

Output format:
(cuisine_type, dietary_restriction, meal_type)
...
"""

# ‡πÄ‡∏ï‡∏¥‡∏° code ‡πÄ‡∏û‡∏∑‡πà‡∏≠ gen tuple ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ 15 ‡∏≠‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô


generated_tuples_text = generate(
    system_prompt="You generate dimension tuples.",
    query=tuple_gen_prompt
)

uniques_tuple = [
    t.strip("- ").strip()
    for t in generated_tuples_text.strip().split("\n")
    if t.strip()
]

print(uniques_tuple)
````

````python
query_prompt = """
You will receive 10 tuples.
Each tuple is structured as:
(cuisine_type, dietary_restriction, meal_type)

Task:
- For each tuple, generate 3 natural user queries (total 30 queries)
- Each query must include all 3 dimensions
- The queries must sound natural and realistic
- Vary phrasing and tone

Output format:

[Tuple: (Thai, vegetarian, dinner)]
1. ...
2. ...
3. ...

[Tuple: (...)]
...
"""
queries_text = generate(
    system_prompt="You generate user queries.",
    query=query_prompt
)

queries = []
for q in queries_text.split('\n'):
    stripped_q = q.strip()

    # ‡∏Å‡∏£‡∏≠‡∏á‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏ß‡πà‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô Header ‡∏Ç‡∏≠‡∏á Tuple (‡∏ó‡∏µ‡πà‡∏Ç‡∏∂‡πâ‡∏ô‡∏ï‡πâ‡∏ô‡∏î‡πâ‡∏ß‡∏¢ '[')
    if stripped_q and not stripped_q.startswith('['):
        # ‡∏Å‡∏£‡∏≠‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏ô‡∏≥‡∏´‡∏ô‡πâ‡∏≤ (‡πÄ‡∏ä‡πà‡∏ô "1. ", "2. ") ‡∏≠‡∏≠‡∏Å‡πÑ‡∏õ
        if stripped_q[0].isdigit() and stripped_q[1:3] in ['. ', '.']:
            # ‡∏•‡∏ö‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÅ‡∏•‡∏∞‡∏à‡∏∏‡∏î‡∏≠‡∏≠‡∏Å‡πÑ‡∏õ
            query_content = stripped_q.split('.', 1)[1].strip()
            queries.append(query_content)
        elif not stripped_q[0].isdigit():
             # ‡∏Å‡∏£‡∏ì‡∏µ‡∏ó‡∏µ‡πà LLM ‡∏ï‡∏≠‡∏ö‡πÇ‡∏î‡∏¢‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÄ‡∏•‡∏Ç‡∏ô‡∏≥‡∏´‡∏ô‡πâ‡∏≤ (‡∏ñ‡πâ‡∏≤‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡∏∂‡πâ‡∏ô)
             queries.append(stripped_q)


print("-" * 50)
print(f"Generated {len(queries)} user queries.")

print(queries)
````

````python
# ‡∏ï‡πâ‡∏≠‡∏á‡∏£‡∏±‡∏ô‡πÇ‡∏Ñ‡πâ‡∏î generate ‡πÅ‡∏•‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ queries (30 queries) ‡∏Å‡πà‡∏≠‡∏ô
# ... ‡πÇ‡∏Ñ‡πâ‡∏î‡∏™‡∏£‡πâ‡∏≤‡∏á queries ...

dataset = []

for q in queries:
    # ‡∏£‡∏±‡∏ô‡∏ö‡∏≠‡∏ó‡πÅ‡∏•‡∏∞‡∏™‡πà‡∏á Trace ‡πÑ‡∏õ LangFuse ‡πÇ‡∏î‡∏¢‡∏≠‡∏±‡∏ï‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥
    actual_output = generate(system_prompt=recipe_bot_prompt, query=q)

    # ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö Open Coding
    dataset.append({
        "query": q,
        "expected_output": "...",
        "actual_output": actual_output,
        "pass_fail": "...",
        "reason": ""
    })

print("Completed running bot. Trace data collected.")
print("-" * 50)
print(f"Total data points in dataset: {len(dataset)}") # ‡∏Ñ‡∏ß‡∏£‡πÄ‡∏õ‡πá‡∏ô 30
````
